### Q.10). Find the top10 channel names having the highest number of videos in the dataset and the bottom10 having lowest number of videos.


```python
# top10 channel names having the highest number of videos in the dataset 
top_10_channels  = data_set.groupby(["channel_title"])["title"].count().sort_values(ascending=False).head(10)
print(f"top_10_channels name is :{top_10_channels }")

# bottom10 channel names having the lowest number of videos in the dataset 
bottom_10_channels  = data_set.groupby(["channel_title"])["title"].count().sort_values(ascending=True).head(10)
print(f"bottom_10_channels names is :{bottom_10_channels}")
```

## Q.11) Find the title of the video which has the maximum number of likes and the title of the video having minimum likes and write your inferences.



```python
#Finding the title of the video having least number of likes. 
video_min_likes    = data_set["likes"].min()
video_min_likes_title = data_set[data_set["likes"] == video_min_likes  ]["title"].values[0]
print(f"Title of the videos with minimum likes: {[video_min_likes_title]}")
```

```python
#Finding the title of the video having least number of likes. 
video_max_likes  = data_set["likes"].max()
video_max_likes_title  = data_set[data_set["likes"] == video_max_likes ]["title"].values[0]
print(f"Title of the videos with maximum likes: {[video_max_likes_title]}")
```

### Inferences:

- "BTS () 'Dynamite' Official MV" has collected the maximum number of likes, suggesting that it is highly popular and well-received by the audience. This could be attributed to the immense popularity of BTS and their strong fanbase, as well as the quality and appeal of the music video itself.

The low number of likes could be due to various factors such as the content of the video, viewer preferences, or the specific audience for this type of content.

## Q.13) Does the number of views have any effect on how many people disliked the video? Support your answer with a metric and a plot.


```python
# Analyze the relationship between views and dislikes
corr = data_set["view_count"].corr(data_set["dislikes"])
corr
```

```python
import matplotlib.pyplot as plot
```

```python
# data
X = data_set["view_count"]
Y = data_set["dislikes"]
plot.scatter(X,Y,c='purple')
plot.xlabel('Views')
plot.ylabel('Dislikes')
plot.title(f"Correlation between Views and Dislikes is : {corr:.2f}")
plot.show()
```

### inferences:
    

- The correlation coefficient between number of views and the number of dislikes is approximately 0.68
- it's indicates that positive relationship between the number of views and the number of dislikes. As the number of views increases, the number   of dislikes tends to increase as well.
